```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python

#Reading the data
home_data = pd.read_csv('/kaggle/input/sample-dataset/home_dataset.csv')
 
#Extracting the features from dataset    
home_prices = home_data['HousePrice'].values 
home_size = home_data['HouseSize'].values 

#Visualise the data
plt.scatter(home_prices, home_size, marker ='o',edgecolors ='red')
plt.title("House prices vs. House size")
plt.xlabel('House Size (sq ft)')
plt.ylabel('House Price ($)')
plt.show()



```

```python
#training data
x_train,x_test,y_train,y_test = train_test_split(home_prices,home_size)

x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)

price_model = LinearRegression()
price_model.fit(x_train,y_train)

predictions = price_model.predict(x_test)

#plotting the data
plt.scatter(x_test,y_test,marker='o',edgecolor='blue',label='Actual price')
plt.title("Property Price Prediction with Linear Regression")
plt.plot(x_test, predictions, color='red', linewidth= 2, label='Predicted price')
plt.xlabel("House Size (sq.ft)")
plt.ylabel("House Price(million $)")
plt.legend()
plt.show()



```

Linear Regression used to predict property prices using a small housing dataset
